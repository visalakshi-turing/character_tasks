```python
import numpy as np
import matplotlib.pyplot as plt

def biased_normal_distribution(input_number):
    # Constants for the distribution
    min_val, max_val = 1, 15
    std_dev = 2.5  # Fixed standard deviation

    # Base mean for input_number 50 (biased towards lower end)
    base_mean = 1.5

    # Adjust the mean based on the input number (linear mapping)
    # The mean shifts towards max_val as the input number increases, and towards min_val as it decreases
    mean = np.interp(input_number, [0, 100], [min_val, max_val])
    
    # Adjust mean for the bias towards lower numbers
    adjusted_mean = (mean + 3*base_mean) / 5

    # Sample from a normal distribution
    while True:
        sample = np.random.normal(adjusted_mean, std_dev)
        if sample >= min_val and sample <= max_val:
            break

    return int(sample)



test_cases = [0, 25, 50, 75, 100]


grand_counts = {}
for case in test_cases:
    outputs = [biased_normal_distribution(case) for _ in range(1000)]
    print(f"Input: {case}, Outputs: {outputs}")

    counts = {}
    for output in outputs:
        if output not in counts:
            counts[output] = 0
        if output not in grand_counts:
            grand_counts[output] = 0
        counts[output] += 1
        grand_counts[output] += 1

    # Plot the distribution
    plt.figure(figsize=(10, 5))
    plt.bar(counts.keys(), counts.values())
    plt.title(f"Input: {case}")
    plt.show()


# Plot the distribution
plt.figure(figsize=(10, 5))
plt.bar(grand_counts.keys(), grand_counts.values())
plt.title(f"Grand Distribution")
plt.show()
```
